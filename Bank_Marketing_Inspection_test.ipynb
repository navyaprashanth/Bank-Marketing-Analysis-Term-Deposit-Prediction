````python
# Create a new binary feature 'was_contacted_in_previous_campaign'
# If pdays is 999, it means 'no', otherwise 'yes'
df['was_contacted_in_previous_campaign'] = df['pdays'].apply(lambda x: 'no' if x == 999 else 'yes')

# Now, let's confirm the new column's value counts
print("\nValue Counts for 'was_contacted_in_previous_campaign':")
print(df['was_contacted_in_previous_campaign'].value_counts())
print("-" * 30)

# You can also drop the original 'pdays' column if desired, or keep it for now.
# For simplicity, we'll drop it after this step, as 'previous' and 'poutcome'
# already capture related information.
df = df.drop('pdays', axis=1)

print("\nColumns after dropping 'pdays':")
print(df.columns)
print("-" * 30)
````

````python
from sklearn.preprocessing import LabelEncoder
import pandas as pd # Ensure pandas is imported if not already

# First, encode the target variable 'y'
# 'no' will typically become 0, 'yes' will become 1 (alphabetical order)
le = LabelEncoder()
df['y'] = le.fit_transform(df['y'])

print("Value counts for 'y' after Label Encoding:")
print(df['y'].value_counts())
print("-" * 30)

# Identify all remaining categorical columns for One-Hot Encoding
# Make sure to include the new 'was_contacted_in_previous_campaign'
categorical_cols_for_ohe = df.select_dtypes(include='object').columns.tolist()

print("\nCategorical columns to be One-Hot Encoded:")
print(categorical_cols_for_ohe)
print("-" * 30)

# Perform One-Hot Encoding
# drop_first=True helps avoid multicollinearity by dropping one category per feature
df = pd.get_dummies(df, columns=categorical_cols_for_ohe, drop_first=True)

print("\nShape of the DataFrame after One-Hot Encoding:")
print(df.shape)
print("-" * 30)

print("\nFirst 5 rows of the DataFrame after encoding (check new columns):")
print(df.head())
print("-" * 30)
````

````python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns that need scaling
# These are the original numerical columns, excluding the target 'y'
# and any one-hot encoded columns (which are already 0s and 1s)
numerical_cols_to_scale = ['age', 'duration', 'campaign', 'previous',
                           'emp.var.rate', 'cons.price.idx', 'cons.conf.idx',
                           'euribor3m', 'nr.employed']

# Create a StandardScaler instance
scaler = StandardScaler()

# Apply scaling to the identified numerical columns in your DataFrame
df[numerical_cols_to_scale] = scaler.fit_transform(df[numerical_cols_to_scale])

print("\nFirst 5 rows of the DataFrame after numerical feature scaling:")
print(df.head())
print("-" * 30)

print("\nDescriptive statistics for scaled numerical columns (check mean near 0, std dev near 1):")
print(df[numerical_cols_to_scale].describe())
print("-" * 30)
````

````python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df.drop('y', axis=1) # All columns except 'y'
y = df['y']             # The target column 'y'

# Split the data into training and testing sets
# test_size=0.3 means 30% of the data will be used for testing
# random_state=42 ensures reproducibility of the split
# stratify=y ensures that the proportion of classes in y is maintained in both splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
print("-" * 30)

print("\nValue counts for y_train:")
print(y_train.value_counts(normalize=True))
print("-" * 30)

print("\nValue counts for y_test:")
print(y_test.value_counts(normalize=True))
print("-" * 30)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the Logistic Regression model
log_reg_model = LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced')
# Using 'liblinear' solver for robustness on smaller datasets (or when L1/L2 regularization is needed)
# class_weight='balanced' is crucial for imbalanced datasets, it automatically
# adjusts weights inversely proportional to class frequencies.
log_reg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_log_reg = log_reg_model.predict(X_test)
y_proba_log_reg = log_reg_model.predict_proba(X_test)[:, 1] # Probability of the positive class (1 = 'yes')

# Evaluate the model
print("--- Logistic Regression Model Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log_reg):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_log_reg):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_log_reg):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_log_reg):.4f}")
print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba_log_reg):.4f}")

print("\nConfusion Matrix:")
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)
print(cm_log_reg)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.title('Confusion Matrix - Logistic Regression')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_log_reg))
print("-" * 50)
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, class_weight='balanced')
# class_weight='balanced' helps with the imbalanced dataset
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1] # Probability of the positive class (1 = 'yes')

# Evaluate the model
print("--- Random Forest Classifier Model Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_rf):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_rf):.4f}")
print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba_rf):.4f}")

print("\nConfusion Matrix:")
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(cm_rf)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.title('Confusion Matrix - Random Forest Classifier')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
print("-" * 50)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get feature importances from the trained Random Forest model
# Make sure rf_model from the previous step is still available
feature_importances = rf_model.feature_importances_

# Create a Series for better visualization and sorting
features_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})
features_df = features_df.sort_values(by='Importance', ascending=False)

# Print the top 15 most important features
print("--- Top 15 Feature Importances (Random Forest) ---")
print(features_df.head(15))
print("-" * 50)

# Visualize the top 15 features
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=features_df.head(15), palette='viridis')
plt.title('Top 15 Feature Importances from Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
````
